```python
3. #Import Libraries (Code Cell)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
4. #Load Dataset (Code Cell)

df = pd.read_csv("monthly_sales.csv")
df.head()
```

🟩 5. Data Understanding (EDA Start)
Markdown
## Data Understanding

We first explore the structure, size, and basic information of the dataset.


```python
df.shape
df.columns
df.info()
df.describe()
```

🟩 6. Data Preparation
Markdown
## Data Preparation

We convert the Date column into datetime format and create additional useful columns.

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
df["Year"] = df["Date"].dt.year
```

```python
7. #Feature Engineering (VERY IMPORTANT)

df["Profit"] = df["Sales"] - df["Expenses"]
df.head()
```

Exploratory Data Analysis (EDA)

📊 A. Monthly Sales Trend
### Monthly Sales Trend

```python
monthly_sales = df.groupby("Date")["Sales"].sum()

fig, ax = plt.subplots()
monthly_sales.plot(ax=ax)

ax.set_title("Monthly Sales Trend")
ax.set_xlabel("Date")
ax.set_ylabel("Sales")

plt.show()
```

📊 B. Sales by Shop

### Total Sales by Shop

```python
shop_sales = df.groupby("Shop")["Sales"].sum()

fig, ax = plt.subplots()
shop_sales.plot(kind="bar", ax=ax)

ax.set_title("Total Sales by Shop")

plt.show()
```

📊 C. Profit Trend

### Profit Trend Over Time

```python
monthly_profit = df.groupby("Date")["Profit"].sum()

fig, ax = plt.subplots()
monthly_profit.plot(ax=ax)

ax.set_title("Monthly Profit Trend")

plt.show()
```

D. Shop Comparison Over Time

### Sales Comparison Across Shops

```python
pivot = df.pivot_table(values="Sales", index="Date", columns="Shop")

fig, ax = plt.subplots()
pivot.plot(ax=ax)

ax.set_title("Sales Comparison by Shop")

plt.show()
```

🟩 9. Key Insights (Markdown Cell)

## Key Insights

- Sales show an overall increasing trend over time
- Some months consistently perform better than others (seasonality)
- Certain shops outperform others consistently
- Profit fluctuates, indicating varying expense management
- Some shops may have high sales but low profit due to high expenses

## Recommendations

* Shops should track monthly sales trends to identify peak periods and plan accordingly.

* Expense management should be improved to increase profitability, especially in months where sales are high but profits remain low.

* Underperforming shops should carefully evaluate their performance while considering additional factors such as product offerings, pricing strategies, and customer demand. Simply copying high-performing shops may not be effective without understanding these underlying differences.

* Location may significantly influence performance. Shops in high-traffic or urban areas may naturally generate more sales than those in less populated areas.

* Seasonal trends should be leveraged for better inventory planning, staffing, and targeted marketing campaigns.

* It is recommended that shops begin collecting additional data (e.g., product categories, customer behavior, and location details) to enable deeper and more accurate analysis in the future.


## Conclusion
